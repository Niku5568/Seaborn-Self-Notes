```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import numpy as np
```

# Categorial plots

### Categorial scatter plot
    - stip plot
    - swarmplot
### Categorial Distibution plot
    - box plot
    - voilin

### Categorial Estimate plot
    - point plot
    - bar plot
    - count plot
    

```python
tips = sns.load_dataset('tips')
tips.head(4)
```

```python
gap=px.data.gapminder()
gap.head(3)
```

```python
# scatter - numeric - numeric values
sns.scatterplot(data=tips,x="total_bill",y="tip")
```

```python
# strip plot
# axis level 
sns.stripplot(data=tips,x="day",y="total_bill")
```

```python
sns.catplot(data=tips,x="day",y="total_bill")
```

```python
sns.catplot(data=tips,x="day",y="total_bill",jitter=0.2,hue="sex")
```

```python
# swarmplot
sns.catplot(data=tips,x="day",y="total_bill",kind="swarm")
```

```python
sns.swarmplot(data=tips,x="day",y="total_bill",hue="sex")
```

# Box Plot

```python
sns.boxplot(data=tips,y="total_bill",x="day")
```

```python
sns.catplot(data=tips,y="total_bill",x="day",kind="box")
```

```python
sns.boxplot(data=tips,y="total_bill",x="day",hue="sex")
```

```python
# single boxplot
sns.boxplot(data=tips,y="total_bill")
```

# violin plot = ( box + kde) 


```python
sns.violinplot(data=tips,x="day",y="total_bill")
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="violin",hue="sex",split=True)
```

# barplot

```python
sns.barplot(data=tips,x="sex",y="total_bill",)
```

```python
sns.catplot(data=tips,x="sex",y="total_bill",
            kind="bar",
            hue="smoker",
            estimator=np.min # to perform mathamatic functions
           )
```

# poinplot 

```python
sns.pointplot(data=tips,x="sex",y="total_bill",
              estimator=np.mean, # to perform mathamatic functions
              hue="smoker"
           )
```

# countplot

```python
sns.countplot(data=tips,x="sex",
              hue="day"
           )
```

# faceting using cat - multiple plots 

```python
sns.catplot(data=tips,x="sex",y="total_bill",kind="box",col="smoker",row="time"
           )
```

# Regplot - reggration plot

```python
sns.regplot(data=tips,x="total_bill",y="tip")
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip",hue="sex")
```

```python
sns.residplot(data=tips,x="total_bill",y="tip")
```

```python
sns.catplot(data=tips,x="sex",y="total_bill",kind="violin",col="day",row="time")
```
