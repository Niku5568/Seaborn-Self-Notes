```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import numpy as np
```

# Relational plots
    - scatterplot
    - lineplot
    - relplot - > to use figure level to plot scatter and lineplots using relplot.

```python
tips = sns.load_dataset('tips')
tips.head(4)
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips,
                hue="sex", # add additional perameter
                style="time", # add another perameters
                size="size", # another permater               
               )
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,
            kind="scatter",
            hue="sex",
            style="time",
            size="size")
plt.show()
```

```python
gap=px.data.gapminder()
gap.head(3)
```

```python
india=gap[gap["country"]=="India"]
india.head(3)
```

```python
sns.lineplot(data=india,x="year",y="lifeExp")
```

```python
sns.relplot(data=india,x="year",y="lifeExp",kind="line")
```

```python
con=gap[gap["country"].isin(["India","Pakistan","China"])]
con.sample(3)
```

```python
sns.relplot(data=con,x="year",y="lifeExp",kind="line",hue="country")
```

```python
sns.lineplot(data=con,x="year",y="lifeExp",hue="country")
```

```python
con=gap[gap["country"].isin(["India","Brazil","Germany"])]
con.sample(3)
```

```python
sns.relplot(data=con,x="year",y="lifeExp",kind="line",hue="country",style="continent",size="continent")
```

# Facet Plot - same like subplots

```python
sns.relplot(x="total_bill",y="tip",data=tips,
            hue="sex", # add additional perameter
            style="time", # add another perameters
            size="size", # another permater   
            kind="scatter",
            col="sex", # multiple column plots on the basis of category
            row="day" # multiple row plots on the basis of category
            
           )
plt.show()
```

```python
sns.relplot(x="lifeExp",y="gdpPercap",data=gap,
            kind="scatter",
            col="year",
            col_wrap=3,
           )
plt.show()
```

# Matrix plot
    heatmap - axis level function
    clustermp - hierarically  - clustermap
    

```python
gap.head(1)
```

```python
gp=gap.pivot_table(index="country",columns="year",values="lifeExp")
gp

gap[gap["continent"]=="Europe"].gap.pivot_table(index="country",columns="year",values="lifeExp")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(gp,annot=True)
plt.show()
```

```python
eu=gap[gap["continent"]=="Asia"].pivot_table(index="country",columns="year",values="lifeExp")
eu
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(eu,annot=True,linewidth=0.5,cmap="jet")
```
